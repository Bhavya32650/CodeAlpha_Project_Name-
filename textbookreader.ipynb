```python
!pip install gTTS==2.3.1
!pip install PyPDF2==3.0.1
from gtts import gTTS
import PyPDF2
import os
def pdf_to_speech(pdf_file_path, output_file):
  """Converts the content of a PDF file to speech and saves it to an audio file.
  Args:
    pdf_file_path: The path to the PDF file.
    output_file: The name of the output audio file.
  """
  with open(pdf_file_path, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ""
    for page_num in range(len(pdf_reader.pages)):
      page = pdf_reader.pages[page_num]
      text += page.extract_text()
  tts = gTTS(text=text, lang='en')
  tts.save(output_file)
  print(f"Audio saved to: {output_file}")
# Example usage with a PDF file:
pdf_file_path = "sampletextbook_fixed.pdf"  # Replace with your PDF file path.
output_file = "textbook_audio.mp3"
pdf_to_speech(pdf_file_path, output_file)
# To play the audio within Colab (optional):
from IPython.display import Audio
Audio(output_file, autoplay=False)
```
